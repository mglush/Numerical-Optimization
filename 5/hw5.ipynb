#### Construct the frequency matrix

```python
# English alphabet
'''
n = 26
A = np.zeros((n,n))
for i in range(len(letters) - 1):
    A[ord(letters[i]) - ord('A'), ord(letters[i + 1]) - ord('A')] += 1 # using the english 'A'
# annnd the last letter is followed by the first
A[ord(letters[-1]) - ord('A'), ord(letters[0]) - ord('A')] += 1
'''
#'''
# Russian alphabet
n = 33
A = np.zeros((n,n))
for i in range(len(letters) - 1):
    letter_1 = ord(letters[i]) - ord('А')
    letter_2 = ord(letters[i + 1]) - ord('А')
    if (letter_1 >= 0 and letter_1 < 33 and letter_2 >= 0 and letter_2 < 33):
        A[letter_1, letter_2] += 1 # using the russian 'А'
# annnd the last letter is followed by the first
A[ord(letters[-1]) - ord('А'), ord(letters[0]) - ord('А')] += 1
#'''
```

#### Check correctness of the frequency matrix 

```python
U, s, Vt = spla.svd(A)
# check the row sums and col sums of the frequency matrix are equal
# they should be equal as it's just the frequency of each letter
# occurring in the text
print((np.sum(A, axis = 0) == np.sum(A, axis = 1)).all())
# compare our singular values with Moler's research paper
for i in range(len(s)):
    print("s[", i, "]:\n", np.round(s[i], 4))
```

#### Rank 2 approximation of the frequency matrix $A$

```python
U, s, Vt = spla.svd(A)
A_rank_2 = 0
for i in range(2):
    A_rank_2 += s[i] * np.outer(U[:,i], Vt[i,:])
# inspect the relative error norm
print(npla.norm(A_rank_2 - A)/npla.norm(A))
```

#### Plot each $i$th letter in the alphabet at coordinates $$(u_{i,\text{ }2}\text{ , }v_{i,\text{ }2}\text{ })$$

```python
x = U[:,1]
y = Vt[1,:]

plt.figure(figsize = (10, 10)) # change size to large
plt.axis('square')

# zoom in and out
limits = (-.65, .65)
plt.xlim(limits)
plt.ylim(limits)

# plot each character in the alphabet
for i in range(33):
    plt.text(x[i], y[i], chr(i + ord('А')))

# x axis and y axis in black
plt.plot(limits, (0, 0), 'black')
plt.plot((0, 0), limits, 'black')

# title and axis labels
# plt.title('The Gettysburg Address: Vowels vs. Consonants')
plt.title('The Declaration of Independence: Vowels vs. Consonants')
# plt.title('Ruslan and Lyudmila: Vowels vs. Consonants')
plt.xlabel('Likeliness To Be Followed By Consonant =>')
plt.ylabel('Likeliness To Be Preceded By Consonant =>')
```
